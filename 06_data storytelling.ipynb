## Takeaway

Corporate customers spend ~23% more on average (₹3,421 vs ₹2,785 for SME) — and this gap is driven by customer segment, not income or city, so targeting customers by segment will be more effective than targeting by income or location.

## Context

Based on 500 cleaned customer records across 7 cities and 3 segments (Retail, Corporate, SME).

## Evidence

1. **Segment predicts spending, income doesn't.** Corporate customers spend ₹3,421 on average vs ₹2,785 for SME — a real, sample-size-backed gap (134 and 133 customers respectively). Meanwhile, the correlation between income and purchase amount is just 0.03 — essentially no relationship.

2. **This isn't just an income story in disguise.** Even though Delhi and Pune have the highest average customer income, the segment effect holds independently — income differences by city don't explain why Corporate customers specifically outspend others.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/customer_data_cleaned.csv")

# Chart 1: Purchase amount by segment
plt.figure(figsize=(6,4))
segment_avg = df.groupby('segment')['purchase_amount'].mean().sort_values(ascending=False)
sns.barplot(x=segment_avg.index, y=segment_avg.values, order=segment_avg.index)
plt.title('Average Purchase Amount by Segment')
plt.ylabel('Average Purchase Amount (₹)')
plt.show()
```

**Corporate customers spend ~23% more than SME (₹3,421 vs ₹2,785)** — segment is a stronger driver of spending than income.

```python
# Chart 2: Income vs Purchase Amount
plt.figure(figsize=(6,5))
sns.scatterplot(data=df, x='income', y='purchase_amount')
plt.title('Income vs Purchase Amount')
plt.show()
```

**Income shows almost no relationship with purchase amount** (correlation ≈ 0.03) — ruling out income as the explanation for the segment-based spending gap above.

## Recommendation

Prioritize the next marketing or loyalty campaign budget toward Corporate-segment customers specifically, rather than splitting spend evenly across segments or targeting by income bracket.

**Why:** Segment — not income — is the real driver of spending (Corporate spends ~23% more than SME, while income shows almost no correlation with spend). A segment-targeted campaign should yield a higher return per rupee spent than an income-based one, since it's targeting the variable that actually predicts spending behavior.

**Expected impact:** Reallocating campaign spend toward the Corporate segment, even without increasing total budget, should improve overall campaign ROI — since the same spend is now aimed at the group most likely to respond with higher purchases.
