```python
import pandas as pd

df = pd.read_csv("data/customer_data_cleaned.csv")
print("Rows loaded:", len(df))
print("Columns:", df.columns.tolist())
```

## Core Metrics — Definitions

1. **Total Customers** — Count of unique, de-duplicated customer records in the cleaned dataset.
2. **Average Income** — Mean of the `income` column, calculated only on non-missing, non-placeholder values (already handled in Task 4 cleaning).
3. **Average Purchase Amount** — Mean of `purchase_amount` across all customers.
4. **Average Satisfaction Score** — Mean of `satisfaction_score` across all customers.
5. **Customer Count by Segment** — Count of customers grouped by `segment` (Retail/Corporate/SME).

## Aggregation Grain

- Customer-level metrics (count, averages): one row per unique customer.
- Segment breakdown: one row per segment (3 total).

```python

df['segment'] = df['segment'].replace({'Sme': 'SME'})

total_customers_rows = len(df)
total_customers_unique_id = df['customer_id'].nunique()

avg_income = df['income'].mean()
avg_purchase = df['purchase_amount'].mean()
avg_satisfaction = df['satisfaction_score'].mean()
segment_counts = df['segment'].value_counts()

print("Total Customers (row-based, recommended):", total_customers_rows)
print("Total Customers (unique ID, known to undercount):", total_customers_unique_id)
print("Average Income: ₹{:,.0f}".format(avg_income))
print("Average Purchase Amount: ₹{:,.0f}".format(avg_purchase))
print("Average Satisfaction Score:", round(avg_satisfaction, 2))
print("\nCustomers by Segment:")
print(segment_counts)
```

```python
# Manual cross-check: does segment count add up to total?
print("Segment counts sum:", segment_counts.sum())
print("Matches total customers?", segment_counts.sum() == total_customers_rows)

# Manual cross-check: recompute average income a different way
manual_avg_income = df['income'].sum() / len(df)
print("Manual avg income check:", round(manual_avg_income, 2))
print("Matches pandas mean?", round(manual_avg_income, 2) == round(avg_income, 2))
```
