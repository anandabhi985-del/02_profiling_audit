```python
import pandas as pd

df = pd.read_csv("data/customer_data_cleaned.csv")
print("Rows loaded:", len(df))
print(df.columns.tolist())
```

```python
# Trend: customers per signup year
customers_by_year = df.groupby('signup_year')['customer_id'].count()
print("Customers by year:\n", customers_by_year)

# Trend: average purchase amount by signup year
purchase_by_year = df.groupby('signup_year')['purchase_amount'].mean()
print("\nAvg purchase amount by year:\n", purchase_by_year)
```

## Headline

**Corporate customers spend 13% above company average, while SME customers spend 8% below it — segment, not income or city, is the strongest signal of customer value.**

## Recommended Action

Shift the next marketing/retention budget allocation to weight Corporate-segment customers more heavily than Retail or SME, rather than splitting spend evenly across all three segments.

**Why:** Corporate customers spend 13% above the company average; SME customers spend 8% below it. Equal-split spending currently under-invests in the segment that returns the most value per customer.

**Expected impact:** Reallocating spend toward Corporate, even at the same total budget, should improve overall campaign ROI, since the same money is now directed at the segment most likely to convert into higher purchase value.

**Secondary flag (not to bury):** Average satisfaction is 2.89/5, with "2" the most common score — below the scale midpoint. This doesn't yet have a root cause identified and deserves a follow-up investigation (e.g. is it concentrated in one segment or city?) before the next reporting period, since unresolved low satisfaction could undercut the spending gains above.
